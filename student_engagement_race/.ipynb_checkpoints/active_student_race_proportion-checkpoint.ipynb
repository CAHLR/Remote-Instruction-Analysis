# How did total number of weekly active students change over the semesterÂ¶

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import numpy as np


# load student by race
f = open('/home/jenny/remote_learning_phase1_deliverable_07202020/user_racial_groups.pkl', 'rb')
stu_race = pickle.load(f)
races = list(stu_race.keys())
print(stu_race.keys())

# Student events include submitting assignment, commenting on submission, creating discussion topic, creating discussion entries.

# load submission
submission = pd.read_csv('/home/jenny/remote_learning_phase1_deliverable_07202020/Spring 2020/submissions/submissions_000', sep='\t', header=0)
submission = submission.loc[(submission['created_at'].notnull())&(submission['created_at']>='2020-01-14')&(submission['created_at']<='2020-05-15'), ['user_id', 'assignment_id', 'created_at','submitted_at']] # no duplicate exists
submission = submission.loc[(submission['submitted_at'].notnull())&(submission['submitted_at']>='2020-01-14')&(submission['submitted_at']<='2020-05-15')]

# loading submission comments
comments = pd.read_csv('/home/jenny/remote_learning_phase1_deliverable_07202020/Spring 2020/submission_comments/submission_comments_000', sep='\t', header=0)

# loading discussions
announce = pd.read_csv('/home/jenny/remote_learning_phase1_deliverable_07202020/Spring 2020/discussion_topics/discussion_topics_000', sep='\t', header=0)
announce = announce.loc[(announce['created_at']>='2020-01-14')&(announce['created_at']<='2020-05-15')]

# loading discussion entries
discussion = pd.read_csv('/home/jenny/remote_learning_phase1_deliverable_07202020/Spring 2020/discussion_entry/discussion_entry_000', sep='\t', header=0)
discussion = discussion.loc[(discussion['created_at']>='2020-01-14')&(discussion['created_at']<='2020-05-15')]
```

```python
# plot all races together
african = stu_race['African American']
asian = stu_race['Asian']
chicano = stu_race['Chicano/Latino']
inter = stu_race['International']
native = stu_race['Native American/Alaskan Native']
pacific = stu_race['Pacific Islander']
white = stu_race['White']
users = [white, asian, inter, chicano, african, native, pacific]


submission_create = submission.loc[:, ['user_id', 'created_at']]
submission_create.columns = ['user_id', 'time']
submission_submit = submission.loc[:, ['user_id', 'submitted_at']]
submission_submit.columns = ['user_id', 'time']
comments = comments.loc[(comments['message_size_bytes']>0)&(comments['created_at']>='2020-01-14')&(comments['created_at']<='2020-05-15'), ['author_id', 'submission_id', 'assignment_id', 'created_at', 'message_size_bytes']]
comments = comments.loc[:, ['author_id', 'created_at']]
comments.columns = ['user_id', 'time']
announce = announce.loc[(announce['workflow_state']=='active'), ['user_id', 'created_at']]
announce.columns = ['user_id', 'time']
discussion = discussion.loc[(discussion['workflow_state']=='active'), ['user_id', 'created_at']]
discussion.columns = ['user_id', 'time']
user_time = pd.concat((submission_create, submission_submit, comments, announce, discussion), axis=0)
# transfer day to week
user_time['time'] = pd.to_datetime(user_time['time'], format='%Y-%m-%d %H:%M:%S.%f')
user_time['time'] = user_time['time'].apply(lambda x: x.date())
user_time['time'] = user_time['time'].apply(lambda x: x.isocalendar()[1])
min_week = min(user_time['time'].tolist())
user_time['time'] = user_time['time'] - min_week+1

active_user_num = []
all_users = user_time['user_id'].drop_duplicates()
for i in range(len(users)):
    active_user_num.append(len(set(all_users).intersection(set(users[i]))))
    
print(active_user_num)

user_time_group = user_time.groupby('time')
user_actions_num = [[],[],[],[],[],[],[]]
for i in user_time_group.groups.keys():
    group = user_time_group.get_group(i) # week
    for j in range(len(users)):
        race_actions = group.loc[group['user_id'].isin(users[j])]
        num_users = len(race_actions['user_id'].drop_duplicates())
        user_actions_num[j].append([i, num_users/active_user_num[j]])
for i in range(len(user_actions_num)):
    user_actions_num[i] = sorted(user_actions_num[i], key=lambda x: x[0])

user_actions_num = np.array(user_actions_num)
user_actions_num = user_actions_num[:, :, 1]
date = ['01-14', '01-21', '01-28', '02-04', '02-11', '02-18', '02-25', '03-03', '03-10', '03-17', '03-24', '03-31', '04-07', '04-14', '04-21', '04-28', '05-05', '05-12']
plt.plot(date, user_actions_num[0], scalex=True, scaley=True, data=None, marker='o', label='White')
plt.plot(date, user_actions_num[1], scalex=True, scaley=True, data=None, marker='x', label='Asian')
plt.plot(date, user_actions_num[2], scalex=True, scaley=True, data=None, marker='+', label='International')
plt.plot(date, user_actions_num[3], scalex=True, scaley=True, data=None, marker='', label='White')

plt.legend()

plt.yticks(fontsize=70)
plt.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 90
fig_size[1] = 50
plt.rcParams["figure.figsize"] = fig_size
plt.title('Weekly active students proportion in Spring 2020', fontsize=100)
plt.xlabel('week', fontsize=70)
plt.ylabel('proportion of active students(%)', fontsize=70)
plt.show()
```
