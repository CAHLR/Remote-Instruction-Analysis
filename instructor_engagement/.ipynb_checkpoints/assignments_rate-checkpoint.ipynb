# Rate of assignments assigned per week over the year (especially in March)

```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import matplotlib.ticker as ticker


# load assignments
assignments = pd.read_csv('../../remote_learning_data_Spring_2020/assignments/assignments_000', sep='\t', header=0)
assignments = assignments.loc[assignments['workflow_state']=='published', ['assignment_id', 'asn_created_at', 'asn_unlock_at', 'asn_due_at']]  # no duplicate exists

# load assignments overrides
assignments_overrides = pd.read_csv('../../remote_learning_data_Spring_2020/assignments_overrides/assignments_overrides_000', sep='\t', header=0)
assignments_overrides = assignments_overrides.loc[assignments_overrides['due_at'].notnull(), ['assignment_id', 'due_at']].drop_duplicates()
assignments = pd.merge(assignments, assignments_overrides, on='assignment_id', how='left')
assignments['new_due_at'] = assignments[['asn_due_at', 'due_at']].apply(lambda x: x['asn_due_at'] if pd.isnull(x['due_at']) else x['due_at'], axis=1)
assignments['asn_created_at'] = pd.to_datetime(assignments['asn_created_at'], format='%Y-%m-%d %H:%M:%S.%f')
assignments['asn_unlock_at'] = pd.to_datetime(assignments['asn_unlock_at'], format='%Y-%m-%d %H:%M:%S.%f')
assignments['new_due_at'] = pd.to_datetime(assignments['new_due_at'], format='%Y-%m-%d %H:%M:%S.%f')
```

```python
create = assignments.loc[(assignments['asn_created_at']>='2020-01-14')&(assignments['asn_created_at']<='2020-05-15'), ['assignment_id', 'asn_created_at']].drop_duplicates()
create['asn_created_at'] = create['asn_created_at'].apply(lambda x: x.date())
create_group = create.groupby('asn_created_at').size()
fig,ax = plt.subplots()

# set x-axis ticks
myFmt = mdates.DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(myFmt)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.set_xlim([datetime.date(2020, 1, 10), datetime.date(2020, 5, 18)])

plt.plot(create_group, linewidth=5.0)
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)
plt.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 90
fig_size[1] = 60
plt.rcParams["figure.figsize"] = fig_size
plt.title('Daily creating rate of assignment', fontsize=100, pad=30)
#plt.xlabel('date', fontsize=65, labelpad=120)
plt.ylabel('number of assignments', fontsize=65)

# annotation on x-axis

ax.annotate('semester began', fontsize=65, xy=(150, -50), xytext=(150, -180), xycoords='axes points', ha='center',
           arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('remote instruction began', color='red', fontsize=65, xy=(2330, -50), xytext=(2330, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='red', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('semester ended', fontsize=65, xy=(4950, -50), xytext=(4950, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('05-15', fontsize=60, xy=(4880, -55), xycoords='axes points')

ax.annotate('', fontsize=65, xy=(2850, -100), xytext=(3000, -100),
            xycoords='axes points',
            arrowprops=dict(arrowstyle='|-|,widthA=0.3,widthB=0.3', linewidth=5))
ax.annotate('spring recess', fontsize=65, xy=(2840, -180), xycoords='axes points')
plt.axvline(x=datetime.date(2020, 3, 10), color = 'r', linewidth=10, alpha=0.35, linestyle='--')
plt.show()

# line plot of number of created assignments per day:
```

```python
unlock = assignments.loc[(assignments['asn_unlock_at']>='2020-01-14')&(assignments['asn_unlock_at']<='2020-05-15'), ['assignment_id', 'asn_unlock_at']].drop_duplicates()
unlock['asn_unlock_at'] = unlock['asn_unlock_at'].apply(lambda x: x.date())
unlock_group = unlock.groupby('asn_unlock_at').size()

fig,ax = plt.subplots()

# set x-axis ticks
myFmt = mdates.DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(myFmt)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.set_xlim([datetime.date(2020, 1, 10), datetime.date(2020, 5, 18)])

plt.plot(unlock_group, linewidth=5.0)
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)
plt.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 90
fig_size[1] = 60
plt.rcParams["figure.figsize"] = fig_size
plt.title('Daily unlocking rate of assignments', fontsize=100, pad=30)
#plt.xlabel('date', fontsize=65, labelpad=120)
plt.ylabel('number of assignments', fontsize=65)

# annotation on x-axis

ax.annotate('semester began', fontsize=65, xy=(150, -50), xytext=(150, -180), xycoords='axes points', ha='center',
           arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('remote instruction began', color='red', fontsize=65, xy=(2330, -50), xytext=(2330, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='red', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('semester ended', fontsize=65, xy=(4950, -50), xytext=(4950, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('05-15', fontsize=60, xy=(4880, -55), xycoords='axes points')

ax.annotate('', fontsize=65, xy=(2850, -100), xytext=(3000, -100),
            xycoords='axes points',
            arrowprops=dict(arrowstyle='|-|,widthA=0.3,widthB=0.3', linewidth=5))
ax.annotate('spring recess', fontsize=65, xy=(2840, -180), xycoords='axes points')
plt.axvline(x=datetime.date(2020, 3, 10), color = 'r', linewidth=10, alpha=0.35, linestyle='--')

plt.show()

# line plot of number of unlocked assignments per day:
```
