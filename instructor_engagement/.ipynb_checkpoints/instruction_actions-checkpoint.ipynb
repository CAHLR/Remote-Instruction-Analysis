# How did total number of daily instructor events change over the semester

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import matplotlib.ticker as ticker


# load assignments
assignments = pd.read_csv('../../remote_learning_data_Spring_2020/assignments/assignments_000', sep='\t', header=0)
assignments = assignments.loc[(assignments['asn_created_at']>='2020-01-14')&(assignments['asn_created_at']<='2020-05-15'), ['assignment_id', 'asn_created_at', 'asn_unlock_at', 'asn_due_at']]  # no duplicate exists
assi_time = assignments['asn_created_at']

# load assignments overrides
assignments_overrides = pd.read_csv('Spring 2020/assignments_overrides/assignments_overrides_000', sep='\t', header=0)
assignments_overrides = assignments_overrides.loc[(assignments_overrides['created_at']>='2020-01-14')&(assignments_overrides['created_at']<='2020-05-15'), ['assignment_id', 'created_at']]
over_time = assignments_overrides['created_at']


# load assign groups
assign_groups = pd.read_csv('Spring 2020/assignment_groups/assignment_groups000', sep='\t', header=0)
assign_groups = assign_groups.loc[(assign_groups['group_created_at']>='2020-01-14')&(assign_groups['group_created_at']<='2020-05-15')]
assi_group_time = assign_groups['group_created_at']

# load grading
submission = pd.read_csv('Spring 2020/submissions/submissions_000', sep='\t', header=0)
submission = submission.loc[(submission['graded_at'].notnull())&(submission['graded_at']>='2020-01-14')&(submission['graded_at']<='2020-05-15'), ['user_id', 'assignment_id', 'graded_at']] # no duplicate exists
grade_time = submission['graded_at']

# loading submission comments
comments = pd.read_csv('Spring 2020/submission_comments/submission_comments_000', sep='\t', header=0)
f = open('user_id_groups.pkl', 'rb')
user = pickle.load(f)
teacher = user['teacher']
ta = user['ta']
users = teacher + ta
comments = comments.loc[(comments['author_id'].isin(users))&(comments['message_size_bytes']>0)&(comments['created_at']>='2020-01-14')&(comments['created_at']<='2020-05-15'), ['author_id', 'submission_id', 'assignment_id', 'created_at', 'message_size_bytes']]
comments_time = comments['created_at']

# loading announcements
announce = pd.read_csv('Spring 2020/discussion_topics/discussion_topics_000', sep='\t', header=0)
announce = announce.loc[(announce['workflow_state']=='active')&(announce['user_id'].isin(users))]
announce = announce.loc[(announce['created_at']>='2020-01-14')&(announce['created_at']<='2020-05-15')]
announce_time = announce['created_at']

#comments['created_at'] = pd.to_datetime(comments['created_at'], format='%Y-%m-%d %H:%M:%S.%f')
discussion_entry = pd.read_csv('Spring 2020/discussion_entry/discussion_entry_000', sep='\t', header=0)
discussion_entry = discussion_entry.loc[(discussion_entry['workflow_state']=='active')&(discussion_entry['created_at']>='2020-01-14')&(discussion_entry['created_at']<='2020-05-15')]
entry_time = discussion_entry['created_at']

time = pd.concat((assi_time, over_time, assi_group_time, grade_time, comments_time, announce_time, entry_time), axis=0)
#unlock = assignments.loc[(assignments['asn_unlock_at']>'2020-01-14')&(assignments['asn_unlock_at']<'2020-05-25'), ['assignment_id', 'asn_unlock_at']].drop_duplicates()
#unlock['asn_unlock_at'] = unlock['asn_unlock_at'].apply(lambda x: x.date())
#unlock_group = unlock.groupby('asn_unlock_at').size()
```

```python
time = pd.to_datetime(time, format='%Y-%m-%d %H:%M:%S.%f')
time = time.apply(lambda x: x.date())
print(time)
```

```python
time1 = time.to_frame(name='date')
print(time1)
time1 = time1.groupby('date').size()
fig,ax = plt.subplots()

# set x-axis ticks
myFmt = mdates.DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(myFmt)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.set_xlim([datetime.date(2020, 1, 10), datetime.date(2020, 5, 18)])

plt.plot(time1, linewidth=5.0)
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)
plt.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 90
fig_size[1] = 60
plt.rcParams["figure.figsize"] = fig_size
plt.title('Daily action rate of instructor groups', fontsize=100, pad=30)
#plt.xlabel('date', fontsize=65, labelpad=120)
plt.ylabel('number of actions', fontsize=65)

# annotation on x-axis

ax.annotate('semester began', fontsize=65, xy=(150, -50), xytext=(150, -180), xycoords='axes points', ha='center',
           arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('remote instruction began', color='red', fontsize=65, xy=(2330, -50), xytext=(2330, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='red', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('semester ended', fontsize=65, xy=(4950, -50), xytext=(4950, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('05-15', fontsize=60, xy=(4880, -55), xycoords='axes points')

ax.annotate('', fontsize=65, xy=(2850, -100), xytext=(3000, -100),
            xycoords='axes points',
            arrowprops=dict(arrowstyle='|-|,widthA=0.3,widthB=0.3', linewidth=5))
ax.annotate('spring recess', fontsize=65, xy=(2840, -180), xycoords='axes points')
plt.axvline(x=datetime.date(2020, 3, 10), color = 'r', linewidth=10, alpha=0.35, linestyle='--')
plt.show()
```
