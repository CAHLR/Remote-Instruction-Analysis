# Did instructors set more assignments overrides after Covid-19? Was the extended period longer on average after Covid-19? 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
import matplotlib.dates as mdates
import datetime
import matplotlib.ticker as ticker


# load assignments
assignments = pd.read_csv('Spring 2020/assignments/assignments_000', sep='\t', header=0)
assignments = assignments.loc[assignments['workflow_state']=='published', ['assignment_id', 'asn_due_at', 'asn_updated_at']]  # no duplicate exists

# calcuate assignments overrides before and after covid
assignments_overrides = pd.read_csv('Spring 2020/assignments_overrides/assignments_overrides_000', sep='\t', header=0)
assignments_overrides = assignments_overrides.loc[assignments_overrides['due_at'].notnull(), ['assignment_id', 'due_at', 'set_type', 'title', 'created_at', 'unlock_at', 'updated_at']]
assignments_overrides['created_at'] = pd.to_datetime(assignments_overrides['created_at'], format='%Y-%m-%d %H:%M:%S.%f')
overrides_before = assignments_overrides.loc[(assignments_overrides['created_at']>'2020-01-14')&(assignments_overrides['created_at']<'2020-03-01')]

# number of overrides per day before and after 2020-03-01
print('number of overrides per day before covid-19: ', len(overrides_before) / ((pd.to_datetime('2020-03-01') - pd.to_datetime('2020-01-14')) / np.timedelta64(1, 'D')))
overrides_after = assignments_overrides.loc[(assignments_overrides['created_at']>'2020-03-01')&(assignments_overrides['created_at']<'2020-05-25')]
print('number of overrides per day after covid-19: ', len(overrides_after) / ((pd.to_datetime('2020-5-25') - pd.to_datetime('2020-03-01')) / np.timedelta64(1, 'D')))

```

```python
assignments_overrides = assignments_overrides.loc[(assignments_overrides['created_at']>='2020-01-14')&(assignments_overrides['created_at']<='2020-05-15')]
assignments_overrides['created_at'] = assignments_overrides['created_at'].apply(lambda x: x.date())
time = assignments_overrides.groupby('created_at').size()
fig,ax = plt.subplots()
# set x-axis ticks
myFmt = mdates.DateFormatter('%m-%d')
ax.xaxis.set_major_formatter(myFmt)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.set_xlim([datetime.date(2020, 1, 10), datetime.date(2020, 5, 18)])

plt.plot(time, linewidth=5.0)
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)
plt.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 90
fig_size[1] = 60
plt.rcParams["figure.figsize"] = fig_size
plt.title('Number of assignment extensions created per week (day) over the semester', fontsize=100, pad=30)
#plt.xlabel('date', fontsize=65, labelpad=120)
plt.ylabel('number of submissions', fontsize=65)

# annotation on x-axis

ax.annotate('semester began', fontsize=65, xy=(150, -50), xytext=(150, -180), xycoords='axes points', ha='center',
           arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('remote instruction began', color='red', fontsize=65, xy=(2330, -50), xytext=(2330, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='red', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('semester ended', fontsize=65, xy=(4950, -50), xytext=(4950, -180), xycoords='axes points', ha='center', arrowprops=dict(facecolor='black', shrink=0.1, headlength=20, headwidth=20))
ax.annotate('05-15', fontsize=60, xy=(4880, -55), xycoords='axes points')

ax.annotate('', fontsize=65, xy=(2850, -100), xytext=(3000, -100),
            xycoords='axes points',
            arrowprops=dict(arrowstyle='|-|,widthA=0.3,widthB=0.3', linewidth=5))
ax.annotate('spring recess', fontsize=65, xy=(2840, -180), xycoords='axes points')
plt.axvline(x=datetime.date(2020, 3, 10), color = 'r', linewidth=10, alpha=0.35, linestyle='--')

plt.show()
```

```python
# delayed time 
assignments_overrides = assignments_overrides.loc[:, ['assignment_id', 'due_at']].drop_duplicates()
assignments_override = pd.merge(assignments, assignments_overrides, on='assignment_id', how='inner')
assignments_override['asn_due_at'] = pd.to_datetime(assignments_override['asn_due_at'], format='%Y-%m-%d %H:%M:%S.%f')
assignments_override['due_at'] = pd.to_datetime(assignments_override['due_at'], format='%Y-%m-%d %H:%M:%S.%f')
assignments_override = assignments_override.loc[assignments_override['asn_due_at'].notnull()]
assignments_override = assignments_override.loc[(assignments_override['asn_due_at']<assignments_override['due_at']) & (assignments_override['asn_due_at']>'2020-01-14') & (assignments_override['due_at']<'2020-05-25')]
assignments_override['delay_time'] = (assignments_override['due_at'] - assignments_override['asn_due_at']) / np.timedelta64(1, 'D')
assignments_before = assignments_override.loc[assignments_override['asn_due_at']<'2020-03-01']
assignments_before_delay = assignments_before['delay_time']
print(assignments_before_delay.describe())
plt.hist(assignments_before_delay, bins=range(1, 50), density=True)
plt.show()
assignments_after = assignments_override.loc[assignments_override['asn_due_at']>='2020-03-01']
assignments_after_delay = assignments_after['delay_time']
print(assignments_after_delay.describe())
plt.hist(assignments_after_delay, bins=range(1, 50), density=True)
plt.show()
```
